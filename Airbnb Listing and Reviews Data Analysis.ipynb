### Question 7: For reviews that are written in Jan, 2016, what are the most popular ten words except stopping words? What about reviews written in March, 2017 and April, 2018?
**Hint**: You need to use NTLK stopping words to find the stopping words in Python. You also need to remove punctuation and numbers

**Question 7 Answer:**\
Jan,2016: ['great','stay','place','clean','location','house','us','nice','would','host']\
March,2017:['great','place','stay','location','clean','would', 'nice','host', 'house', 'comfortable']\
April,2018:['great','place', 'stay', 'location', 'clean', 'would', 'host', 'nice', 'la', 'house']

```python
# You need to provide the code to answer the previous quesiton
import string
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
from collections import Counter


df_reviews=pd.read_csv("reviews/reviews.csv")
df_reviews["comments"]=df_reviews["comments"].str.lower()
df_reviews.dropna(subset=['comments'],inplace=True)
df_reviews['date'] = pd.to_datetime(df_reviews['date']) 
df_reviews = df_reviews.set_index('date')
df_rev=df_reviews['2016-01-01':'2016-01-31']   #'2017-03-01':'2017-03-31' #'2018-04-01':'2018-04-30'
df_rev.reset_index(inplace=True)
df_rev
commentlist=[]
for t in range(len(df_rev)):
    commentlist.append(df_rev.loc[t,"comments"])

    

rev_punctuation=[]
for m in range(len(commentlist)):
    trantab = str.maketrans({key: None for key in string.punctuation})
    j = commentlist[m].translate(trantab)
    rev_punctuation.append(j)
hh=' '.join(rev_punctuation)
gg=hh.replace('’',"")

stop_words = set(stopwords.words('english'))
word_tokens = word_tokenize(gg)
filtered_sentence = [w for w in word_tokens if not w in stop_words]
filtered_sentence = []
for w in word_tokens:
    if w not in stop_words:
        filtered_sentence.append(w)

N = 10
print(Counter(filtered_sentence).most_common(N))



   
  
```
